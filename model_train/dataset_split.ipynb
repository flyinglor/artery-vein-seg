```python
import pandas as pd
from sklearn.model_selection import train_test_split, KFold

# Read full dataset
df = pd.read_excel("./dataset/metadata.xlsx")

# 1. Split out test set first (e.g., 20%)
df_trainval, df_test = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)

# 2. Prepare K-Fold on the remaining data
kf = KFold(n_splits=5, shuffle=True, random_state=42)

folds = []
for fold_idx, (train_idx, val_idx) in enumerate(kf.split(df_trainval)):
    train_df = df_trainval.iloc[train_idx]
    val_df = df_trainval.iloc[val_idx]

    folds.append({
        "fold": fold_idx + 1,
        "train": train_df,
        "val": val_df
    })

    print(f"Fold {fold_idx + 1}: Train size = {len(train_df)}, Validation size = {len(val_df)}")

# Save test set separately
df_test.to_excel("./dataset/split/test_set.xlsx", index=False)

# Save folds
for fold in folds:
    fold["train"].to_excel(f"./dataset/split/train_fold_{fold['fold']}.xlsx", index=False)
    fold["val"].to_excel(f"./dataset/split/val_fold_{fold['fold']}.xlsx", index=False)

```
